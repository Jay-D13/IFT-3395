```python
from tqdm import tqdm
```

```python
from solution import *

x_train, y_train, x_test, y_test = load_data()
```

```python
results = {}

for C in tqdm([1, 5, 10]):
    print(f'Computing losses for C: {C}')

    svm = SVM(eta=0.0001, C=C, niter=200, batch_size=100, verbose=False)
    train_losses, train_accs, test_losses, test_accs = svm.fit(x_train, y_train, x_test, y_test)

    results[C] = {
        'train_losses': train_losses,
        'train_accs': train_accs,
        'test_losses': test_losses,
        'test_accs': test_accs,
    }
```

```python
import matplotlib.pyplot as plt
```

```python
label_to_string = {
    'train_losses': "Perte points d'entraînement",
    'train_accs': "Exactitude des points d'entraînement",
    'test_losses': 'Perte points de test',
    'test_accs': "Exactitude des points test",
}

for label in results[1].keys():
    x = range(1, 201)
    y1 = results[1][label]
    y2 = results[5][label]
    y3 = results[10][label]

    plt.plot(x, y1, marker='', color='black', linewidth=1, label="C=1")
    plt.plot(x, y2, marker='', color='purple', linewidth=1, label="C=5")
    plt.plot(x, y3, marker='', color='red', linewidth=1, label="C=10")
    plt.title(label_to_string[label])
    plt.xlabel("Epochs")
    plt.ylabel("Rate")
    plt.legend()
    plt.show()
```
